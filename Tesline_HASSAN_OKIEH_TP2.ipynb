```python
#Exercice 2
# Affiche quelques doublons pour vérification
print(df[df.duplicated()])
```

```python
# Supprimer les doublons exacts
df = df.drop_duplicates()
```

```python
# Vérifier qu'ils ont bien disparu
print("Doublons ensuite:", df.duplicated().sum())
```

```python
#Exercice 4
# 1. Vérifier le type de Start Year
print(df['Start Year'].dtype)
```

```python
# 2. Montrer les valeurs problématiques
print(df[df['Start Year'].str.contains("Year|AD", na=False)]['Start Year'].head())
```

```python
# 3. Nettoyage : retirer le texte et convertir en numérique
df['Start Year'] = df['Start Year'].str.replace(r"[^\d]", "", regex=True)
df['Start Year'] = pd.to_numeric(df['Start Year'], errors='coerce').astype('Int64')

print(df['Start Year'].head())
```

```python
#Exercice 4

# 1 Region : supprimer les lignes nulles
df = df.dropna(subset=['Region'])


```

```python
# 2 Event Name : remplacer nulls par « Non nommé »
df['Event Name'] = df['Event Name'].fillna("Non nommé")
```

```python
# 3 Total Deaths : imputer par 0 (ou autre logique)
df['Total Deaths'] = df['Total Deaths'].fillna(0)
```

```python
# 5 Vérifier le résultat
print(df.isnull().sum())
```

```python
#Exercice 5

# 1. Identifier les valeurs négatives dans Total Deaths
neg_deaths = df[df["Total Deaths"] < 0]
print("Valeurs négatives Total Deaths :")
print(neg_deaths)
```

```python
# 2. Identifier les valeurs aberrantes dans Magnitude (> 10)
outliers_mag = df[df["Magnitude"] > 10]
print("\nValeurs aberrantes Magnitude :")
print(outliers_mag)
```

```python
# 3. Décision de traitement

#  Les morts négatifs sont impossibles → suppression
#  Magnitude >10 est irréaliste → suppression

```

```python
# 4. Appliquer le traitement
df = df[df["Total Deaths"] >= 0]
df = df[df["Magnitude"] <= 10]
```

```python
# 5. Vérification du résultat
print("\nAprès nettoyage :")

print("Valeurs négatives restantes :",
      df[df["Total Deaths"] < 0].shape[0])

print("Magnitudes >10 restantes :",
      df[df["Magnitude"] > 10].shape[0])
```

```python
# Exercice 6 

# 1. Nettoyer Region → enlever espaces + format titre
df["Region"] = df["Region"].str.strip().str.title()
```

```python
#2.Vérifier
print("Catégories Region :")
print(df["Region"].value_counts())

print("Nombre de catégories Region :", df["Region"].nunique())
```

```python
#Il y a maintenant 5 catégories pour la variable Region : Asia, Americas, Europe, Africa et Oceania.
```

```python
# 3. Nettoyer Disaster Type
df["Disaster Type"] = df["Disaster Type"].str.strip().str.title()
```

```python
# Vérifier
print("\nCatégories Disaster Type :")
print(df["Disaster Type"].value_counts())
```

```python
# 4. Nettoyer Country
df["Country"] = df["Country"].str.strip().str.title()
```

```python
# 5 . Corriger variantes de pays
df["Country"] = df["Country"].replace({
    "Usa": "United States",
    "United States Of America": "United States",
    "U.s.a.": "United States",
    "Etats-Unis": "United States",
    "Uk": "United Kingdom",
    "U.k.": "United Kingdom"
})
```

```python
# Vérification finale
print("\nPays uniques après correction :")
print(df["Country"].value_counts())
```

```python
# EXERCICE 7 

# 1. Créer colonne Decennie
df["Decennie"] = (df["Start Year"] // 10) * 10
```

```python
# 2. Créer colonne Has_Deaths
df["Has_Deaths"] = df["Total Deaths"] > 0
```

```python
# 3.  Vérification
print(df[["Start Year", "Decennie", "Total Deaths", "Has_Deaths"]].head())
```

```python
# EXERCICE 8 
# 1. Sélectionner les colonnes utiles pour l'analyse
df_export = df[[
    "Start Year",
    "Decennie",
    "Country",
    "Region",
    "Disaster Type",
    "Total Deaths",
    "Has_Deaths"
]]
```

```python
# 2. Renommer les colonnes pour qu'elles soient claires et sans espaces
df_export = df_export.rename(columns={
    "Start Year": "Year",
    "Decennie": "Decade",
    "Disaster Type": "Disaster_Type",
    "Total Deaths": "Total_Deaths",
    "Has_Deaths": "Has_Deaths"
})
```

```python
# 3. Vérification finale
print(df_export.info())
print(df_export.head())
```

```python
# 4. Exporter en CSV
df_export.to_csv("catnat_clean.csv", index=False)
```
